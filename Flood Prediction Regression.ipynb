plot_numerics(flood_data):
This function creates visualization pairs (histograms and box plots) for each numeric column in the flood data:


It iterates through all columns in the dataset
For each column, it creates a figure with two subplots side by side (using figsize=(16, 5))
Left subplot: Creates a histogram using seaborn's histplot

Shows the distribution of values
Includes a KDE (Kernel Density Estimation) line
Uses skyblue color
Adds proper titles and labels


Right subplot: Creates a box plot using seaborn's boxplot

Shows the median, quartiles, and potential outliers
Uses lightgreen color
Adds proper titles and labels


Uses tight_layout() to prevent subplot overlap
Shows the plot


get_count_plot(cat):
This function creates a count plot for categorical variables:


Creates a bar plot showing the frequency of each category
Orders the categories by frequency (descending)
Adds labels to each bar showing:

The absolute count
The percentage of total (in parentheses)


Sets a title indicating which categorical variable is being plotted

```python
def plot_numerics(flood_data):
    numeric_columns = flood_data.columns
    # Plot histograms and box plots for each numeric column
    for column in numeric_columns:
        # Create separate subplots for each column
        _, ax = plt.subplots(1, 2, figsize=(16, 5))
        ax = ax.flatten()

        # Histogram
        sns.histplot(flood_data[column], bins=50, kde=True, color='skyblue', ax=ax[0])
        ax[0].set_title(f'Histogram of {column}', fontsize=15, fontweight='bold')
        ax[0].set_xlabel(column, fontsize=12)
        ax[0].set_ylabel('Frequency', fontsize=12)

        # Boxplot
        sns.boxplot(x=flood_data[column], color='lightgreen', ax=ax[1])
        ax[1].set_title(f'Box plot of {column}', fontsize=15, fontweight='bold')
        ax[1].set_xlabel(column, fontsize=12)

        # Show the plots for the current column
        plt.tight_layout()
        plt.show()

        
def get_count_plot(cat):
    ax=sns.countplot(x=flood_data[cat],
                   order=flood_data[cat].value_counts(ascending=False).index);
    abs_values = flood_data[cat].value_counts(ascending=False)
    rel_values = flood_data[cat].value_counts(ascending=False, normalize=True).values * 100
    lbls = [f'{p[0]} ({p[1]:.0f}%)' for p in zip(abs_values, rel_values)]

    ax.bar_label(container=ax.containers[0], labels=lbls)
    ax.set_title("Distribution of "+cat+" Values",fontsize=16)
```

This code performs advanced feature engineering on flood-related data to enhance its predictive capabilities. The first part creates a list of unique values found in numerical columns across both training and testing datasets. The main function, `getFeats`, then transforms the data by creating new features that capture complex interactions and statistical properties of the existing variables.

The function creates two interaction features that model how climate and anthropogenic factors combine, as well as how infrastructure and prevention measures interact. It then calculates various statistical measures across the numerical columns, including basic statistics like mean, standard deviation, and sum, as well as more complex metrics like skewness, kurtosis, and various percentiles. The function also computes advanced statistical measures such as harmonic and geometric means, z-scores, coefficient of variation, and entropy.

Additionally, it creates count-based features for specific values that appear less than 16 times in the dataset. Finally, the function standardizes all numerical columns using StandardScaler, which transforms the features to have zero mean and unit variance. This comprehensive feature engineering approach aims to capture both simple and complex relationships in the data, potentially improving the performance of subsequent flood prediction models.

```python
 unique_vals = []
for df in [X_train, X_test]:
    for col in num_cols:
        unique_vals += list(flood_data[col].unique())

unique_vals = list(set(unique_vals))
#
def getFeats(flood_data):
    
    scaler = StandardScaler()
    
    flood_data['ClimateAnthropogenicInteraction'] = (flood_data['MonsoonIntensity'] + flood_data['ClimateChange']) * (flood_data['Deforestation'] + flood_data['Urbanization'] + flood_data['AgriculturalPractices'] + flood_data['Encroachments'])
    flood_data['InfrastructurePreventionInteraction'] = (flood_data['DamsQuality'] + flood_data['DrainageSystems'] + flood_data['DeterioratingInfrastructure']) * (flood_data['RiverManagement'] + flood_data['IneffectiveDisasterPreparedness'] + flood_data['InadequatePlanning'])
    
    flood_data['sum'] = flood_data[num_cols].sum(axis=1)
    flood_data['std']  = flood_data[num_cols].std(axis=1)
    flood_data['mean'] = flood_data[num_cols].mean(axis=1)
    flood_data['max']  = flood_data[num_cols].max(axis=1)
    flood_data['min']  = flood_data[num_cols].min(axis=1)
    flood_data['mode'] = flood_data[num_cols].mode(axis=1)[0]
    flood_data['median'] = flood_data[num_cols].median(axis=1)
    flood_data['skew'] = flood_data[num_cols].skew(axis=1)
    flood_data['kurt'] = flood_data[num_cols].kurt(axis=1)

    for i in range(10,100,10):
        flood_data[f'{i}th'] = flood_data[num_cols].quantile(i/100, axis=1)
        
    flood_data['harmonic'] = len(num_cols) / flood_data[num_cols].apply(lambda x: (1/x).mean(), axis=1)
    flood_data['geometric'] = flood_data[num_cols].apply(lambda x: x.prod()**(1/len(x)), axis=1)
    flood_data['zscore'] = flood_data[num_cols].apply(lambda x: (x - x.mean()) / x.std(), axis=1).mean(axis=1)
    flood_data['cv'] = flood_data['std'] / flood_data['mean']
    flood_data['Skewness_75'] = (flood_data[num_cols].quantile(0.75, axis=1) - flood_data[num_cols].mean(axis=1)) / flood_data[num_cols].std(axis=1)
    flood_data['Skewness_25'] = (flood_data[num_cols].quantile(0.25, axis=1) - flood_data[num_cols].mean(axis=1)) / flood_data[num_cols].std(axis=1)
    flood_data['2ndMoment'] = flood_data[num_cols].apply(lambda x: (x**2).mean(), axis=1)
    flood_data['3rdMoment'] = flood_data[num_cols].apply(lambda x: (x**3).mean(), axis=1)
    flood_data['entropy'] = flood_data[num_cols].apply(lambda x: -1*(x*np.log(x)).sum(), axis=1)
    
    for v in unique_vals:
        if v<16:
            flood_data['cnt_{}'.format(v)] = (flood_data[num_cols] == v).sum(axis=1)
    
    flood_data[num_cols] = scaler.fit_transform(flood_data[num_cols])
    
    return flood_data
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score
```

This code implements a cross-validation training framework for a LightGBM model to ensure robust flood prediction performance. The function uses 5-fold cross-validation, meaning it splits the data into five parts and trains five different models, each using a different subset as validation data.

For each fold, the function trains a LightGBM model with carefully tuned hyperparameters, including GPU acceleration for faster training. The model uses gradient boosting decision trees with 2000 potential estimators, though early stopping is implemented to prevent overfitting (training stops if performance doesn't improve for 70 rounds). The model incorporates various parameters to control tree growth, feature sampling, and regularization.

During training, the function tracks both training and validation performance using R-squared scores, which measure how well the model's predictions match the actual values. It accumulates predictions for both the validation set and test set, with test predictions being averaged across all folds. This averaging helps create more stable and reliable predictions by combining the insights from multiple models trained on different data splits.

The function returns three key pieces of information: the validation scores from each fold (to assess model stability), predictions for the validation data, and averaged predictions for the test data. This comprehensive validation approach helps ensure the model will generalize well to new, unseen flood data.

```python
## Function for Cross Validation
def cross_val_train(X,y,X_test,params):
    
    spl=5
    test_preds = np.zeros((len(X_test)))
    val_preds = np.zeros((len(X)))
    val_scores, train_scores = [],[]
    
    cv = KFold(spl, shuffle=True, random_state=42)
    
    for fold, (train_ind, valid_ind) in enumerate(cv.split(X,y)):
        
        X_train = X.iloc[train_ind]
        y_train = y[train_ind]
        X_val = X.iloc[valid_ind]
        y_val = y[valid_ind]
        
        model = lgb.LGBMRegressor(boosting_type = 'gbdt', 
                                   n_estimators=2000, 
                                   learning_rate =  0.012, 
                                   device='gpu',
                                   num_leaves = 250, 
                                   subsample_for_bin= 165700, 
                                   min_child_samples= 114, 
                                   reg_alpha= 2.075e-06, 
                                   reg_lambda= 3.839e-07, 
                                   colsample_bytree= 0.9634,
                                   subsample= 0.9592, 
                                   max_depth= 10,
                                   random_state=0,
                                  verbosity=-1
                                 )
        model.fit(X_train, y_train,
                    eval_set=[(X_val, y_val)],
                    callbacks=[lgb.early_stopping(stopping_rounds=70), lgb.log_evaluation(100)]) 
        
        y_pred_trn=model.predict(X_train)
        y_pred_val=model.predict(X_val)
        train_r2 = r2_score(y_train, y_pred_trn)
        val_r2 = r2_score(y_val, y_pred_val)
        print("Fold:",fold, " Train R2:",np.round(train_r2,5), " Val R2:",np.round(val_r2,5))
        
        test_preds += model.predict(X_test[feats])/spl
        val_preds[valid_ind] = model.predict(X_val)
        val_scores.append(val_r2)
        print("-"*50)

    return val_scores, val_preds, test_preds
```

```python
val_scores, val_preds, test_preds=cross_val_train(X,y,X_test,params)
```
